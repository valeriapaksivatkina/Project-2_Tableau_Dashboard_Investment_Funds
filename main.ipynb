Risk calculation

```python
from scipy.stats import norm 
```

```python
df_performance = pd.read_csv('../valeria/data/performance.csv')
df_performance
```

```python
confidence_level = 0.95  
time_period = 1 

std_dev_per_fund = df_performance.groupby('FundID')['Return'].std().fillna(0)  


def calculate_var(std_dev):
    if isinstance(std_dev, float): 
        std_dev_clean = std_dev  
    else:
        std_dev_clean = std_dev.dropna().astype(float)  
    return norm.ppf(1 - confidence_level, loc=0, scale=std_dev_clean) * np.sqrt(time_period)

var_per_fund = std_dev_per_fund.apply(calculate_var)

print("VaR for each fund:")
print(var_per_fund)
```

```python
std__dev_per_fund = std_dev_per_fund.to_frame(name='std_dev_per_fund')
std__dev_per_fund
```

```python
#std_dev_per_fund.reset_index(inplace=True)
#var_per_fund.reset_index(inplace=True)

# Fusionar las series al DataFrame df_performance por FundID
df_performance = pd.merge(df_performance, std_dev_per_fund, on='FundID', how='left')
df_performance = pd.merge(df_performance, var_per_fund, on='FundID', how='left')

# Renombrar las columnas según sea necesario
df_performance.rename(columns={'std_dev_per_fund': 'std_dev_por_fund', 'var_per_fund': 'var_por_fund'}, inplace=True)
```

```python
df_performance.rename(columns={'Return_x': 'Return', 'Return_y': 'std_dev_per_fund', 'Return': 'var_per_fund'}, inplace=True)
```

```python
df_performance
```

```python
# function for defining risk levels 
def determine_risk_level(std_dev_per_fund, var_per_fund):
    if std_dev_per_fund < 0.05 and var_per_fund < 0.1:  # low risk level
        return 'Low'
    elif std_dev_per_fund > 0.1 or var_per_fund > 0.2:  # high risk level
        return 'High'
    else:
        return 'Medium' # medium risk level 
```

```python
df_performance['Risk_level'] = df_performance.apply(lambda row: determine_risk_level(row['std_dev_per_fund'], row['var_per_fund']), axis=1)

df_performance
```
